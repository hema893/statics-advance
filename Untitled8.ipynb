```python
13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
```

```python
The **Central Limit Theorem (CLT)** states that the **sampling distribution of the sample mean** becomes approximately **normal**, regardless of the population's distribution, **as the sample size becomes large** (usually $n \geq 30$).

### Importance:

* Allows use of **normal distribution** for inference (e.g., confidence intervals, hypothesis tests) even when data isn't normally distributed.
* Forms the foundation for many statistical methods.
```

```python
14. How does the Central Limit Theorem relate to the normal distribution?
```

```python
The **Central Limit Theorem (CLT)** relates to the **normal distribution** by stating that:

As the **sample size increases**, the **distribution of the sample mean** approaches a **normal distribution**, **even if the original data is not normal**.

This connection allows statisticians to use **normal distribution-based methods** (like Z-tests or confidence intervals) on sample means from any population, given a sufficiently large sample.
```

```python
15.What is the application of Z statistics in hypothesis testing?
```

```python
**Z statistics** is used in hypothesis testing to determine how far a sample statistic (like the sample mean) is from the population mean in terms of **standard deviations (Z-score)**.

### Applications:

* Test population **means** when the **population standard deviation is known**.
* Compare **sample proportion** to a population proportion.
* Used in **Z-tests** to decide whether to **reject the null hypothesis**.
```

```python
16.How do you calculate a Z-score, and what does it represent?
```

```python
A **Z-score** is calculated using the formula:

$$
Z = \frac{X - \mu}{\sigma}
$$

Where:

* $X$ = data point
* $\mu$ = mean
* $\sigma$ = standard deviation

### What it represents:

The Z-score tells you **how many standard deviations** a data point is from the mean.

* **Positive Z**: above the mean
* **Negative Z**: below the mean
* **Z = 0**: exactly at the mean
```

```python
21. What are the assumptions for applying the Central Limit Theorem?
```

```python
The **Central Limit Theorem (CLT)** applies under these key assumptions:

1. **Random Sampling**: Data should be collected through random sampling.
2. **Independence**: Observations must be independent of each other.
3. **Sample Size**: Sample size should be large (typically **n ≥ 30**); larger if the population is skewed.
4. **Finite Variance**: The population must have a finite mean and variance.

> Under these assumptions, the sampling distribution of the sample mean approaches a **normal distribution**, regardless of the population's shape.
```

```python
def bernoulli_pdf(x, p):
    """
    Calculates the PDF of a Bernoulli distribution.

    Parameters:
    x (int): Outcome value (0 or 1)
    p (float): Probability of success (0 ≤ p ≤ 1)

    Returns:
    float: Probability of observing outcome x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")
    
    return p if x == 1 else 1 - p
```

```python
import numpy as np
from scipy.stats import norm

def z_test(sample_data, mu_0, sigma, alpha=0.05):
    """
    Performs a one-sample Z-test.

    Parameters:
    sample_data (list or array): Sample observations
    mu_0 (float): Hypothesized population mean
    sigma (float): Known population standard deviation
    alpha (float): Significance level (default = 0.05)

    Returns:
    dict: Z-statistic, p-value, and test conclusion
    """
    n = len(sample_data)
    sample_mean = np.mean(sample_data)
    
    # Calculate Z-statistic
    z = (sample_mean - mu_0) / (sigma / np.sqrt(n))
    
    # Two-tailed p-value
    p_value = 2 * (1 - norm.cdf(abs(z)))
    
    # Conclusion
    if p_value < alpha:
        conclusion = "Reject the null hypothesis (significant difference)."
    else:
        conclusion = "Fail to reject the null hypothesis (no significant difference)."
    
    return {
        "Sample Mean": sample_mean,
        "Z-statistic": z,
        "p-value": p_value,
        "Conclusion": conclusion
    }

# === Example Usage ===

# Simulate a sample dataset
np.random.seed(42)
sample = np.random.normal(loc=52, scale=10, size=50)

# Known population parameters
mu_0 = 50      # Hypothesized population mean
sigma = 10     # Known population standard deviation
alpha
```

```python
17. Create a confidence interval for a dataset using Python and interpret the result?
```

```python
import numpy as np
from scipy.stats import norm

def confidence_interval(data, confidence=0.95, sigma=None):
    """
    Calculates a confidence interval for the population mean using Z-distribution.

    Parameters:
    data (list or array): Sample data
    confidence (float): Confidence level (default = 0.95)
    sigma (float or None): Known population standard deviation; if None, use sample std

    Returns:
    tuple: (lower_bound, upper_bound, margin_of_error)
    """
    n = len(data)
    mean = np.mean(data)

    if sigma is None:
        sigma = np.std(data, ddof=1)

    z_critical = norm.ppf(1 - (1 - confidence) / 2)
    se = sigma / np.sqrt(n)
    margin_of_error = z_critical * se

    lower = mean - margin_of_error
    upper = mean + margin_of_error

    return lower, upper, margin_of_error

# === Example Usage ===

# Simulated sample dataset
np.random.seed(0)
sample_data = np.random.normal(loc=100, scale=15, size=50)

# Calculate 95% confidence interval
lower, upper, moe = confidence_interval(sample_data, confidence=0.95)

print(f"Sample Mean: {np.mean(sample_data):.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
print(f"Margin of Error: {moe:.2f}")
```

```python
18. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean?
```

```python
import numpy as np
from scipy.stats import norm

def calculate_confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for the mean of normally distributed data.

    Parameters:
    data (list or array): The dataset
    confidence (float): The confidence level (default is 0.95)

    Returns:
    tuple: (mean, lower_bound, upper_bound, margin_of_error)
    """
    n = len(data)
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # Sample standard deviation

    z_critical = norm.ppf(1 - (1 - confidence) / 2)
    standard_error = std / np.sqrt(n)
    margin_of_error = z_critical * standard_error

    lower = mean - margin_of_error
    upper = mean + margin_of_error

    return mean, lower, upper, margin_of_error

# === Generate sample data ===
np.random.seed(42)
sample_data = np.random.normal(loc=75, scale=12, size=50)  # N(75, 12^2)

# === Calculate confidence interval ===
mean, lower, upper, moe = calculate_confidence_interval(sample_data, confidence=0.95)

# === Print Results ===
print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
print(f"Margin of Error: {moe:.2f}")
```

```python
19.Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution?
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0      # Mean
sigma = 1   # Standard deviation

# Generate a range of x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calculate the PDF
pdf = norm.pdf(x, loc=mu, scale=sigma)

# Plot the PDF
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, color='blue', label=f'N({mu}, {sigma}²)')
plt.fill_between(x, pdf, color='lightblue', alpha=0.5)
plt.title('Probability Density Function (Normal Distribution)')
plt.xlabel('x')
plt.ylabel('Density')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```
